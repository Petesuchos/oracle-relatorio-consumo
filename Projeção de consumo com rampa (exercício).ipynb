```python
import pandas as pd
import numpy as np
```

```python
start_date = '2021-01-01'
end_date = '2021-12-31'
ramp_days = 90
amount = 100_000
```

```python
idx = pd.date_range(start_date, end_date)
```

```python
idx
```

```python
ramp_linspace = np.linspace(start=0.01, stop=1, num=ramp_days)
```

```python
ramp_linspace
```

```python
ramp_array = np.append(ramp_linspace, np.ones(len(idx) - ramp_days))
```

```python
ramp_array
```

```python
amount_per_day = amount / len(idx)
amount_per_day
```

```python
consumption_forecast = amount_per_day * ramp_array
```

```python
df = pd.DataFrame({'Amount' : consumption_forecast, 'Customer': 'ACME Inc.', 'ForecastType': 'Upside'}, index=idx)
```

```python
df
```
