```python
import pandas as pd
```

```python
previsao_df = pd.read_excel('Previsão de Consumo.xlsx', index_col='Data')
```

```python
previsao_df
```

```python
df = previsao_df.groupby(['Cliente', 'Oportunidade', 'Tipo Forecast', pd.Grouper(freq='M')])['Consumo Projetado USD'].sum().to_frame().reset_index()
```

```python
df
```

```python
pivot = df.pivot(index=['Cliente', 'Oportunidade', 'Tipo Forecast'], columns='Data', values='Consumo Projetado USD').fillna(0)
```

```python
pivot
```

```python
pivot.columns
```

```python
from datetime import datetime
def extract_month(datestr):
    dateobj = datetime.strptime(str(datestr), '%Y-%m-%d %H:%M:%S')
    return dateobj.strftime("%B")

    
```

```python
pivot.rename(columns=extract_month, inplace=True)
```

```python
pivot['Total'] = pivot.sum(axis=1)
```

```python
pivot
```

```python
pivot / 1000
```

```python
pivot['Total'].sum()
```
